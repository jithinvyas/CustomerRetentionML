# **PROBLEM STATEMENT**

"Predicting the   retail customer retention potential level as high/medium/low &  Customer Segmentation & Pattern extraction

It is important for any retail company to focus on customer though continuous engagement which translates to increased sales and resultant profitability measured over a period. The customer churn results in not just loss of potential revenue and future profits together with sunk acquisition costs but also negative publicity and incurrence of additional cost of new acquisitions.

On the other hand, Customer focus leads to retention which in turn leads to generation of value, not just volumes. Thus, understanding of customer churn and the underlying factors is central to customer relationship management.

Retailers develop metric to capture customer response towards the product, service, price and competition so that they offer better products and service delivery to customers in comparison with competition. In retail business, a customer is treated as churned once his/her transactions/interactions/visits outdate a particular amount of time. It happens when the product/service is no more used or he uses a product/service of competitor. It is useful to understand defected customer in terms of interests and preferences  to win them back.

It is possible to understand the customer based on historical data of customers containing the characteristics of customers, their transactions, the interactions with them  etc. Customer churn predictions help the retail company devise suitable marketing programs to target specific groups of retail customers as part of continued engagement. To ensure that such marketing spend is appropriately directed to the customers likely to be retained and/or with high probability of retention, you are required to classify the retention potential of customer as High/Medium/Low.


You are expected to classify each customer  as high/medium/low under the target attribute “RententionPotential”. Here identifying the medium level customers is important as this category  indicates where some intelligent action may lead to high possibility of retaining the customers who are likely to churn.

You are also required to segment the customers from the standpoint of customer retention for strategic purposes and extract the top 5 patterns for each level in target attribute, while answering other questions too cited below.

## Time taken by the customer to get interested

```python
CustData=pd.merge(Customerdata_Train,Emaildata_Train,on=['CustomerID'])             ## Merging customerdata and emaildata
CustData=CustData[['CustomerID','DateOfemail','DateOfRegistration','EmailType']]    ## Considering only required features
CustData.drop(CustData[CustData['EmailType']=='WelcomeEmail'].index,inplace=True)   ## Dropping welcome emails
# As the date of welcome email is equal to date of registration we remove the records that have welcome emails.
CustData.head()
```

```python
CustData.info() # Checking for the datatypes and number of records
```

Date of Email is in object format. Let's convert that into the datetime format

```python
CustData["DateOfemail"]=pd.to_datetime(CustData["DateOfemail"]) # Converting into required datatype
```

```python
# Sorting the data by DateofEmail and DateOfRegistration in descending order
CustData1 = CustData.sort_values(by=['CustomerID','DateOfemail','DateOfRegistration'],ascending=False,ignore_index=True)
CustData1.head()
```

```python
CustData2 = CustData1.drop_duplicates(subset=['CustomerID'],keep='first')  ## Maximum time the customer is towards the business
CustData3 = CustData1.drop_duplicates(subset=['CustomerID'],keep='last')   ## Minimum time taken by customer getting intrested towards this business
```

```python
CustData2.shape,CustData3.shape # Checking whether CustData2 and CustData3 have same shape or not
```

```python
# subtracting DateOfRegistration from DateOfemail
CustData2['MaxDays'] = CustData2['DateOfemail'] - CustData2['DateOfRegistration']
CustData3['MinDays'] = CustData3['DateOfemail'] - CustData3['DateOfRegistration']
```

```python
CustData2.head() # Maximum no. of days the customer has been interested/happy with this business
```

```python
CustData3.head()  # Number of days the customer has taken to get interested towards the business
```

```python
# As we have customerID's of the ones who got only Customized emails, let's create a dataframe with all customerID's
tempdf=pd.DataFrame({'CustomerID': Customerdata_Train['CustomerID']}) # Creating a temperory dataframe with all customer ID's
```

```python
# Merging the above results with CustomerID
tempdf1=pd.merge(CustData2[['CustomerID','MaxDays']],tempdf, on='CustomerID',how='right')
tempdf2=pd.merge(CustData3[['CustomerID','MinDays']],tempdf1, on='CustomerID',how='right')
```

```python
# Filling the NaN values with 0 days
tempdf3 = tempdf2.fillna('0 days')
tempdf3.head()
```

**Note:** If MinDays=MaxDays=0 days, it means the customer has not received Customized email

```python
tempdf3.info() # Checking for the datatypes and null values
```

```python
tempdf3[["MinDays","MaxDays"]] = np.int64(tempdf3[["MinDays","MaxDays"]]/np.timedelta64(1,'D'))  # Converting the object into integer by removing 'days'
tempdf3['Days_Spent'] = tempdf3['MaxDays'] - tempdf3['MinDays']  # Total no. of days the customer is positive towards the business
tempdf3.head()
```

```python
# Lineplot from matplotlib library
def lineplot(x_axis,y_axis,data,xlabel,ylabel,title):
  fig,ax = plt.subplots(figsize=(13,6))
  for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(12)
  plt.plot(x_axis,y_axis,data=data,marker='o')
  plt.xlabel(xlabel, fontsize=15, color='black')
  plt.ylabel(ylabel, fontsize=15, color='black')
  fig.suptitle(title, fontsize=22, color='darkblue')
  plt.show()
```

```python
# Lineplot from seaborn library
def snslineplot(x_axis,y_axis,data,xlabel,ylabel,title,target=None,sort=None):
  fig,ax = plt.subplots(figsize=(13,6))
  for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(12)
  sns.lineplot(x_axis,y_axis,data=data,marker='o',markersize=8,hue=target,linewidth=2,sort=sort,ci=None)
  plt.xlabel(xlabel, fontsize=15, color='black')
  plt.ylabel(ylabel, fontsize=15, color='black')
  fig.suptitle(title, fontsize=22, color='darkblue')
  plt.show()
```

```python
def barplot(x_axis,y_axis,data,xlabel,ylabel,title):
  fig,ax = plt.subplots(figsize=(13,6))
  for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(12)
  plt.bar(x_axis,y_axis,data=data, width=0.6)
  plt.xlabel(xlabel, fontsize=15, color='black')
  plt.ylabel(ylabel, fontsize=15, color='black')
  fig.suptitle(title, fontsize=22, color='darkblue')
  plt.show()
```

## Highly correlated features

```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j]) > threshold:
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

# MODEL BUILDING

### Functions

```python
def model_scores(model):
  results = pd.DataFrame()
  results['Train_Recall'] = recall_score(y_train, model.predict(X_train),average='weighted'),
  results['Test_Recall'] = recall_score(y_test, model.predict(X_test),average='weighted'),

  results['Train_Precision'] = precision_score(y_train, model.predict(X_train),average='weighted'),
  results['Test_Precision'] = precision_score(y_test, model.predict(X_test),average='weighted'),

  results['Train_Accuracy'] = accuracy_score(y_train, model.predict(X_train)),
  results['Test_Accuracy'] = accuracy_score(y_test, model.predict(X_test))
  return results
```
